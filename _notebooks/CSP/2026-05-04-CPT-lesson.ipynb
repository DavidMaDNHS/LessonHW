        (function(){
          const codeArea = document.getElementById('runner-code');
          const output = document.getElementById('runner-output');
          const runBtn = document.getElementById('run-code');
          const resetBtn = document.getElementById('reset-code');
          const defaultCode = codeArea?.value || '';

          function safeRun(code) {
            try {
              const logs = [];
              const consoleProxy = { log: (...args) => logs.push(args.map(String).join(' ')) };
              new Function('console', code)(consoleProxy);
              return logs.length ? logs.join('
') : '[No output]';
            } catch (err) {
              return 'Error: ' + err.message;
            }
          }

          if (runBtn && codeArea && output) {
            runBtn.addEventListener('click', () => {
              output.textContent = 'Running...';
              output.textContent = safeRun(codeArea.value);
            });
          }
          if (resetBtn && codeArea && output) {
            resetBtn.addEventListener('click', () => {
              codeArea.value = defaultCode;
              output.textContent = 'Code reset. Press Run to execute.';
            });
          }

          const draftButtons = document.querySelectorAll('.draft-save');
          draftButtons.forEach(btn => {
            btn.addEventListener('click', () => {
              const key = btn.dataset.key;
              const textarea = document.getElementById('draft-' + key);
              const status = document.getElementById('draft-status-' + key);
              if (!key || !textarea) return;
              localStorage.setItem('cpt-draft-' + key, textarea.value);
              if (status) {
                status.textContent = 'Saved.';
                setTimeout(() => status.textContent = '', 2200);
              }
            });
          });
          document.querySelectorAll('.draft-text').forEach(textarea => {
            const key = textarea.id?.replace(/^draft-/,'');
            if (!key) return;
            textarea.value = localStorage.getItem('cpt-draft-' + key) || '';
          });
        })();
